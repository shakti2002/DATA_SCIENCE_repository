```python
import pandas as pd
India_weather= pd.DataFrame({"city":['mumbai','delhi','banglore'],
                           "humdity":[21, 42, 15],
                           "temperature":[32, 40, 6]})
India_weather
```

```python
US_weather= pd.DataFrame({"city":['new york','chicago','orlando'],
                           "humdity":[22, 41, 13],
                           "temperature":[15, 31, 10]})
US_weather
```

```python
df= pd.concat([India_weather,US_weather])
df
```

```python
df= pd.concat([India_weather,US_weather], ignore_index=True)
df
```

```python
df= pd.concat([India_weather,US_weather], keys=["india","us"])
df
```

```python
df.loc["us"]
```

```python
windspeed_df=pd.DataFrame({"city":['delhi','mumbai'],
                           "windspeed":[21, 42]}, index=[1,0])
windspeed_df
```

```python
temperature_df=pd.DataFrame({"city":['mumbai','delhi','banglore'],
                           "temperature":[32, 40, 6]}, index=[0,1,2])
temperature_df
```

```python
df=pd.concat([temperature_df, windspeed_df], axis=0)
df
```

```python
df=pd.concat([temperature_df, windspeed_df], axis=1)
df
```
